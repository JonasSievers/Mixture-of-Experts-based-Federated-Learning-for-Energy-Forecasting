# Merge Weather and Load Data to One Dataset

```python
# Required libraries
import pandas as pd
```

```python
# Read the weather data.
weather = pd.read_csv('../2012-2013 Weather_preprocessed.csv')
weather.head()
```

```python
# Read the load data
load = pd.read_csv('../2012-2013 Load_preprocessed.csv')
load.head()
```

```python
# Check the shape
print('Weather data shape: ',weather.shape)
print('Load data shape: ',load.shape)
```

```python
# Merge 

# Merge the weather and load dataframes
merged_df = pd.merge(load, weather, left_on='Date', right_on='time')

# Remove the redundant "time" column
merged_df.drop('time', axis=1, inplace=True)
```

```python
merged_df.head()
```

```python
# Check the shape

print('Merged_df shape: ',merged_df.shape)
```
