```python
# Set time period
start = datetime(2012, 6, 30)# 2012-07-01 01:00:00
end = datetime(2013, 7, 1)
```

```python
# Get hourly data
df = Hourly('94755', start, end)
df = df.fetch()
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.reset_index()

df['time'] = pd.to_datetime(df['time']) + timedelta(hours=10)

df = df.set_index('time')

# Resample to 1-hour interval and forward fill missing values
df_resampled = df.resample('1H').ffill()

# Reset the index to make the time column a regular column again
df_resampled = df_resampled.reset_index()

df_resampled.head(10)
```

```python
# Matching timeseries with the load dataset. 

weather_sliced = df_resampled[12:8772] 
weather_sliced.shape
```

```python
# Drop the unnecessary columns

weather_sliced = weather_sliced.drop(['prcp', 'snow', 'wpgt', 'tsun', 'coco'], axis=1)
```

```python
#Save the dataframe

weather_sliced.to_csv('weather_final.csv', index=False)
```
