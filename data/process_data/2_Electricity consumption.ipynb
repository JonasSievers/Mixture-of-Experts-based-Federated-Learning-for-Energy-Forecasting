```python
# Required libraries

import pandas as pd
from datetime import timedelta
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the data

load = pd.read_csv('final_dataset_2012-2013_home electricity data.csv')
load.head()
```

```python
# Pivot the dataframe
pivot_df = load.pivot(index='Date', columns='User', values='Consumption_kW').reset_index()

# Rename the columns
pivot_df.columns.name = None

load_final = pivot_df

# Print the resulting dataframe
load_final.head()
```

```python
# Save the dataframe

load_final.to_csv('load_final.csv', index=False)
```
