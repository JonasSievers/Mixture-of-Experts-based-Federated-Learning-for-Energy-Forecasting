```python
# Required libraries

import pandas as pd
from datetime import timedelta
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the weather data.

weather = pd.read_csv('weather_final.csv')
weather.head()
```

```python
# Read the load data

load = pd.read_csv('load_final.csv')
load.head()
```

```python
# Check the shape

print('Weather data shape: ',weather.shape)
print('Load data shape: ',load.shape)
```

```python
# Merge 

# Merge the weather and load dataframes
merged_df = pd.merge(load, weather, left_on='Date', right_on='time')

# Remove the redundant "time" column
merged_df.drop('time', axis=1, inplace=True)
```

```python
merged_df.head()
```

```python
# Check the shape

print('Merged_df shape: ',merged_df.shape)
```

```python
# Save final dataset

merged_df.to_csv('final_dataset.csv', index=False)
```
