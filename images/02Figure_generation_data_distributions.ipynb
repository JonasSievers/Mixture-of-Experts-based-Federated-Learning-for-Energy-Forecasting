```python
#Imports
import pandas as pd
from joypy import joyplot
import matplotlib.pyplot as plt
import os

import sys  
sys.path.append("../")  
```

```python
cwd = os.path.normpath(os.path.dirname(os.getcwd()))
```

# Ridge Plots

```python
pv = pd.read_csv(cwd+'/data/3final_data/Final_PV_dataset.csv', index_col='Date')
pv.index = pd.to_datetime(pv.index)
pv_plot_df = pv[["User1", "User2", "User3", "User4", "User5", "User6", "User7", "User8", "User9","User10"]]

goss_load = pd.read_csv(cwd+'/data/3final_data/Final_Grossload_dataset.csv', index_col='Date')
goss_load.index = pd.to_datetime(goss_load.index)
goss_load_plot_df = goss_load[["User1", "User2", "User3", "User4", "User5", "User6", "User7", "User8", "User9","User10"]]

total_load = pd.read_csv(cwd+'/data/3final_data/Final_Totalload_dataset.csv', index_col='Date')
total_load.index = pd.to_datetime(total_load.index)
total_load_plot_df = total_load[["User1", "User2", "User3", "User4", "User5", "User6", "User7", "User8", "User9","User10"]]

prosumption = pd.read_csv(cwd+'/data/3final_data/Final_Prosumption_dataset.csv', index_col='Date')
prosumption.index = pd.to_datetime(prosumption.index)
prosumption_plot_df = prosumption[["User1", "User2", "User3", "User4", "User5", "User6", "User7", "User8", "User9","User10"]]
```

```python
import seaborn as sns

dfs = [pv_plot_df,
       goss_load_plot_df,
       total_load_plot_df,
       prosumption_plot_df]

# Create subplots
fig, axes = plt.subplots(1, 4, figsize=(20, 6))

# Plotting for each DataFrame
for i, df in enumerate(dfs):
    sns.kdeplot(data=df, ax=axes[i], fill=True, common_norm=False)
    axes[i].set_title(f'DataFrame {i + 1}')

# Set common y-axis label
fig.text(0.5, 0.04, 'PV Generation', ha='center', va='center')

plt.suptitle('Kernel Density Plots of PV Generation', y=1.02)
plt.show()
```

```python
# Ridgeline Plot of PV Generation Density Distribution
fig, axes = joyplot(
    pv_plot_df,
    figsize=(15, 3),
    colormap=plt.cm.autumn_r,
    alpha=0.6,
    linewidth=1.5,
)

plt.title('Ridgeline Plot of PV Generation Density Distribution')
plt.show()
```
